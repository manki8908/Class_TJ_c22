```python
import cv2
import matplotlib.pyplot as plt 

img = cv2.imread('d:/images/apple.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(img.shape)
plt.imshow(img,cmap='gray')
```

```python
import tensorflow as tf 
import numpy as np 

img2 = img/255

img2 = np.expand_dims(img2, axis=0)
img2 = np.expand_dims(img2, axis=3)
print(img2.shape)
img3 = tf.keras.layers.Conv2D(1, 3, activation='relu', input_shape=img2.shape)(img2)
print(img3.shape)
```

```python
# 
pool1 = tf.keras.layers.MaxPool2D(pool_size=(3, 3), strides=(3, 3))(img3)
print(pool1.shape)

result1=np.squeeze(pool1, axis=0)

print(result1.shape)
```

```python
pool2 = tf.keras.layers.AveragePooling2D(pool_size=(3, 3), strides=(3, 3))(img3)
print(pool2.shape)

result2=np.squeeze(pool2, axis=0)

print(result2.shape)
```

```python
plt.imshow(result1,cmap='gray')
```
