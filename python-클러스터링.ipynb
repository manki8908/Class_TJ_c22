```python
#    

#inertia value :          

#   

import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

clusters = range(2, 11)

inertias = []

for n in clusters:

    #  

    model = KMeans(n_clusters=n,random_state=10)

    model.fit(X_train)

    inertias.append(model.inertia_)

    

plt.rcParams["font.size"]=15

plt.plot(clusters, inertias, "-o")

plt.ylabel("inertia")

plt.xlabel("number of clusters")

plt.show()

#(elbow) 

#inertias      3

# 3    
```

```python
#    

#           

# ai : i      

# bi : i        

# (bi - ai) / max(ai,bi)

#    

#           

from sklearn.metrics import silhouette_score

silhouette_score(X_test, pred)
```

```python
from sklearn.cluster import KMeans

#           

#       

#  1   

#             

# k-means++ , random  

model = KMeans(init="k-means++", n_clusters=10, random_state=0)

model.fit(digits.data)

y_pred = model.labels_

print(model.cluster_centers_) #
```

```python
import matplotlib.pyplot as plt

def show_digits(images, labels):

    f = plt.figure(figsize=(8, 2))

    i = 0

    while (i < 10 and i < images.shape[0]):

        ax = f.add_subplot(1, 10, i + 1)

        ax.imshow(images[i], cmap=plt.cm.bone)

        ax.set_title(labels[i])

        ax.xaxis.set_ticks([])

        ax.yaxis.set_ticks([])

        plt.tight_layout()

        i += 1

        

def show_cluster(images, y_pred, cluster_number):

    images = images[y_pred == cluster_number]

    y_pred = y_pred[y_pred == cluster_number]

    show_digits(images, y_pred)
```

```python
for i in range(10):

    show_cluster(digits.images, y_pred, i)



#        
```

```python
from sklearn.metrics.cluster import silhouette_score

# 1.0   ,     
silhouette_score(digits.data, y_pred)
```

```python
#      

#  64   2  

from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X = pca.fit_transform(digits.data)

plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap=plt.cm.Set1)

plt.show()
```

```python
import time

import numpy as np

from sklearn import datasets

from sklearn.cluster import MiniBatchKMeans

from sklearn.feature_extraction.image import extract_patches_2d

faces = datasets.fetch_olivetti_faces()
```

```python
faces.data.shape # 64*64
```

```python
plt.imshow(faces.data[30].reshape(64,64), cmap=plt.cm.gray)  

plt.xticks(())  

plt.yticks(())  

plt.show()
```

```python
#    1024

kmeans = MiniBatchKMeans(n_clusters=81, random_state=0, verbose=True)

#  

patch_size = (20, 20)

buffer = []

t0 = time.time()

# 400  6  : 2400

index = 0

for _ in range(6): #   6 

    for img in faces.images: # 400 

        data = extract_patches_2d(img, patch_size, max_patches=50,random_state=0)

        data = np.reshape(data, (len(data), -1))

        buffer.append(data)

        index += 1

        if index % 10 == 0:

            # 

            data = np.concatenate(buffer, axis=0)

            data -= np.mean(data, axis=0)

            data /= np.std(data, axis=0)

            kmeans.partial_fit(data)

            buffer = []

        if index % 100 == 0:

            print('  : %4i / %i' % (index, 6 * len(faces.images)))

dt = time.time() - t0

print(':', dt)
```

```python
plt.figure(figsize=(10, 10))

for i, patch in enumerate(kmeans.cluster_centers_):

    plt.subplot(9, 9, i + 1)

    plt.imshow(patch.reshape(patch_size), cmap=plt.cm.gray)

    plt.xticks(())

    plt.yticks(())

plt.show()
```

```python
# DBSCAN(Density-Based Spatial Clustering of Applications with Noise)  
#    

#          

#     x  n      

#      

# (   ),   (     )
```

```python
from sklearn.cluster import DBSCAN

import numpy as np

X = np.array([[1, 2], [2, 2], [2, 5],

              [8, 7], [8, 10], [25, 80], [27,78], [40,50]])

# 3,  2

# eps(epsilon) :    ,  0.5

# min_samples :        ,  5

model = DBSCAN(eps=3, min_samples=2).fit(X)

print(model)

plt.scatter(X[:,0],X[:,1],c=model.labels_)

print(model.labels_) #3 (-1   , )
```

```python
def plot_DBSCAN(title, X, eps):

    model = DBSCAN(eps=eps, min_samples=5)

    model.fit_predict(X)

    plt.scatter(X[model.labels_ == 0, 0], X[model.labels_ == 0, 1])

    plt.scatter(X[model.labels_ == 1, 0], X[model.labels_ == 1, 1])

    plt.xticks(())

    plt.yticks(())

    plt.title(title)

plt.figure(figsize=(10, 5))

plt.subplot(121)

plot_DBSCAN("circle", X1, 0.1)

plt.subplot(122)

plot_DBSCAN("moon", X2, 0.1)
```

```python
import pandas as pd

df=pd.read_csv('./DATA/iris.csv')

df.columns
```

```python
cols=df.columns[:-2]

X=df[cols]

y=df['Species']
```

```python
import matplotlib.pyplot as plt

from sklearn.pipeline import make_pipeline

from sklearn.preprocessing import StandardScaler

from sklearn.cluster import DBSCAN

scaler = StandardScaler()

model = DBSCAN(eps=0.5,min_samples=5)

pipeline = make_pipeline(scaler,model)

pred = pd.DataFrame(pipeline.fit_predict(X))

pred.columns=['predict']

result = pd.concat([X,y,pred],axis=1)

result
```
